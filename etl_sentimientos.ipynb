```python
def classify_comment(comment):
    if comment is None:
        return 'No message'
    else:
        sentiment = TextBlob(comment).sentiment.polarity
        if sentiment > 0:
            return 'Positive'
        elif sentiment < 0:
            return 'Negative'
        else:
            return 'Neutral'
```

```python
df_rese√±a['feeling'] = df_rese√±a['text'].apply(lambda x: classify_comment(x) if pd.notnull(x) else 'No message')
```

```python
df_rese√±a['feeling_resp'] = df_rese√±a['resp'].apply(lambda x: classify_comment(x) if pd.notnull(x) else 'No message')
```
